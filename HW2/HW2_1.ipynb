```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
df = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
df
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.replace({'—': np.nan})
df
```

```python
df.dtypes
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
df.dtypes
```

```python
df.isnull().sum() / len(df)
```

```python
df =  df.drop([27], axis=0)
df
```

```python
df = df.fillna(df.mean(numeric_only=True,axis=0))
df
```

```python
df.columns = df.columns.astype(str)
df.columns = df.columns.str.strip()
df.columns
```

```python
average_2019 = df['2019'].mean()
regions_above_average = df[df['2019'] > average_2019]
regions_above_average
```

```python
df['2019'] = pd.to_numeric(df['2019'], errors='coerce')
df
```

```python
highest_birth_rate_2014 = df[df['2014'] == df['2014'].max()]
highest_birth_rate_2014
```

```python
Years = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']
birth_rate = [20.4, 18.9, 15.6, 12.3, 7.3, 12.2, 12.1, 8.0]
plt.plot(Years, birth_rate, marker='o', color='red')
plt.title('Народжуваність за роки у Київській обл.')
plt.xlabel('Рік', color='blue')
plt.ylabel('Народжуванності', color='blue')
plt.grid(True)
plt.show()
```

```python
df.plot(kind='bar', x='Регіон', y='2019', legend=False)
plt.title('Народжуванність по регіонах України у 2019 році')
plt.ylabel('Коефіцієнт народжуваності')
plt.show()
```

```python
Years = ['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019']
birth_rate = [20.4, 18.9, 15.6, 12.3, 7.3, 12.2, 12.1, 8.0]
plt.hist(birth_rate, bins=5, alpha=0.7, color='green', edgecolor='black')
plt.title('Розподіл народжуваності')
plt.xlabel('Народжуваність')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```
