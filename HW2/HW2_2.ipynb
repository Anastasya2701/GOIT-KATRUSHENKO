```python
df_col = df.isna().any()
col_not_del = ['Мова програмування']
df_col[col_not_del] = False
df.drop(df_col.index[df_col], axis=1, inplace=True)  
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df
```

```python
df.shape
```

```python
python_data = df[df["Мова програмування"] == "Python"]
python_data
```

```python
python_data.shape
```

```python
df.groupby('Посада')
df
```

```python
grouped_data = df.groupby(['Посада']).agg({'Зарплата на місяць': ['min', 'max']})
grouped_data
```

```python
def fill_avg_salary(row):
    return row.mean()
grouped_data['avg'] = grouped_data.apply(fill_avg_salary, axis=1)
grouped_data
```

```python
grouped_data['avg'].describe()
```

```python
grouped_data.to_csv('2017_jun_final_1.csv', encoding='utf-8')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.hist(df['Зарплата на місяць'].dropna(), bins=40, color='skyblue', edgecolor='black')
plt.title('Гістограма заробітної плати')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Кількість')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.hist(df['Загальний досвід роботи'].dropna(), bins=30, color='blue', edgecolor='black')
plt.title('Гістограма загального досвіду роботи')
plt.xlabel('Загальний досвід роботи (роки)')
plt.ylabel('Кількість')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(df['Загальний досвід роботи'].dropna(), df['Зарплата на місяць'].dropna(),'o',color='red')
plt.title('Залежність зарплати від досвіду роботи')
plt.xlabel('Загальний досвід роботи (роки)')
plt.ylabel('Зарплата на місяць')
plt.grid(True)
plt.show()
```
